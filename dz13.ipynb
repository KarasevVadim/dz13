```python
movies = pd.read_csv('ml-latest-small/ratings.csv')
movies.head()
```

```python
count_c=movies.groupby('userId').count()[['movieId']].reset_index()
count_min=movies.groupby('userId').min()[['timestamp']].reset_index()
count_min.rename(columns ={'timestamp':'min'}, inplace =True)
count_max=movies.groupby('userId').max()[['timestamp']].reset_index()
aver_t=count_c.merge(count_max,how='left').merge(count_min,how='left')
aver_t['time_']=aver_t['timestamp']-aver_t['min']
aver_t=aver_t[(aver_t['movieId']>100)]
```

```python
aver_t.head()
```

```python
np.mean(aver_t['time_'])
```
